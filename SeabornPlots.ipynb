```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

# SeaBorn Plots:

# 1. Relational Plots:

# Kind:
    1.Scatter
    2.Line

```python
sns.relplot("tip","total_bill",data=df,hue="sex",size="size",kind="scatter")
```

# 2. Categorical Plots:

# Kind:
    “point”, “bar”, “strip”, “swarm”, “box”, “violin”, or “boxen”.

```python
sns.catplot("size","total_bill",data=df,hue="sex",kind="box")
plt.tight_layout()
```

# 3. Distribution plots:

```python
sns.pairplot(df)
```

# 4. Regression Plots:

```python
sns.lmplot("tip","total_bill",data=df,hue="smoker")
```

# 5. Matrix Plots:

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.clustermap(df.corr(),annot=True)
```
